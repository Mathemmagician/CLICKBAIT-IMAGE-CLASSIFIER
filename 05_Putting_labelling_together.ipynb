```python
def generate_labels(df, n=5):
    # n is the number of how many you want to label
    ids = df[df['label'].isna()][:n]['video_id']

    annotations = annotate(
      ids,
      options=['Clickbait', 'Not_Clickbait', 'Maybe'],
      display_fn=lambda filename: display(Image(f'data/imgs/{filename}.jpg'))
    )

    return annotations
```

```python
def apply_labels_and_save(df, annotations):
    # This takes in the array of tuples generated by generate_labels() and applies the 
    # pandas magic to merge new labels into our df
    
    df['label'] = df['video_id'].map(dict(annotations)).fillna(df['label'])
    df.to_pickle("./data/df_partial_labels.pkl")
```
